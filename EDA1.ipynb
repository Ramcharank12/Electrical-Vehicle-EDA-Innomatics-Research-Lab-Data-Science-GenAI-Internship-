```python
# Import required libraries for EDA and visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
df = pd.read_csv('E:\PROJECT-AI-ML\Innomatics Research Lab - Data Science Gen AI\Electric_Vehicle_EDA\dataset.csv')

# Set visual style for the plots
sns.set_style('whitegrid')
```

Univariate Analysis
Univariate analysis focuses on the distribution of individual columns. We will create several visualizations for a single feature to understand its distribution and unique characteristics.

1. Histogram
Let's plot the histogram for the Electric Range to visualize its distribution.

Explanation: The histogram shows the frequency of different electric range values in the dataset. The KDE line helps visualize the data's probability density.

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Electric Range'], kde=True, color='skyblue')
plt.title('Distribution of Electric Range')
plt.xlabel('Electric Range')
plt.ylabel('Frequency')
plt.show()
```

Box Plot
We use a box plot to identify outliers in the Base MSRP column.

Explanation: The box plot displays the quartiles and median of the Base MSRP values. Any points outside the whiskers are considered potential outliers.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Base MSRP'], color='lightgreen')
plt.title('Box Plot of Base MSRP')
plt.xlabel('Base MSRP')
plt.show()
```

Count Plot
Count plots are useful for understanding the frequency of categories. Lets visualize the count of vehicles by Make.

Explanation: The count plot reveals the number of vehicles for each manufacturer, highlighting the most and least common vehicle brands.

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Make', data=df, order=df['Make'].value_counts().index, palette='viridis')
plt.title('Count of Vehicles by Make')
plt.xlabel('Number of Vehicles')
plt.ylabel('Make')
plt.show()
```

Pie Chart
Create a pie chart to visualize the proportion of different Model Year categories.

Explanation: The pie chart shows the percentage distribution of different model years, helping us understand the dataset's age distribution.

```python
plt.figure(figsize=(10, 15))
df['Model Year'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#66c2a5', '#fc8d62', '#8da0cb'])
plt.title('Distribution of Model Year')
plt.ylabel('')
plt.show()
```

KDE Plot
KDE (Kernel Density Estimate) plots are used to visualize the distribution of a continuous variable.

 Explanation: The KDE plot is another way to understand the distribution of the Electric Range variable.

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(df['Electric Range'], shade=True, color='orange')
plt.title('KDE Plot of Electric Range')
plt.xlabel('Electric Range')
plt.show()
```

Bivariate Analysis
Bivariate analysis explores relationships between two variables, usually a combination of a categorical and a numerical variable.

1. Bar Plot
Create a bar plot to see the average Base MSRP by Make.

Explanation: The bar plot shows the average MSRP for each make, giving insights into which brands tend to have higher base prices.

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Make', y='Base MSRP', data=df, palette='coolwarm')
plt.title('Average Base MSRP by Make')
plt.xlabel('Make')
plt.ylabel('Average Base MSRP')
plt.xticks(rotation=90)
plt.show()
```

Plot a scatter plot to see the relationship between Electric Range and Base MSRP.

Multivariate Analysis
Multivariate analysis involves more than two variables to uncover deeper patterns.

1. Heatmap Correlation Matrix
Use a heatmap to see the correlation between numerical features.

Explanation: The heatmap shows how various numerical features relate to each other. Values close to 1 or -1 indicate strong positive or negative correlations.

```python
from sklearn.preprocessing import LabelEncoder

# Create a copy of the DataFrame to work on
encoded_df = df.copy()

# Apply Label Encoding to categorical columns
for column in encoded_df.select_dtypes(include=['object']).columns:
    encoded_df[column] = LabelEncoder().fit_transform(encoded_df[column].astype(str))

# Generate the correlation matrix after encoding
plt.figure(figsize=(12, 10))
sns.heatmap(encoded_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix with Encoded Categorical Features')
plt.show()
```

2. Pair Plot
Create a pair plot to explore pairwise relationships between numerical variables.

Explanation: Pair plots show scatter plots for every pair of numerical features, providing an overview of potential linear relationships and clusters.

```python
print('Pair Plot of Electric Range, Base MSRP, and Model Year')
sns.pairplot(df[['Electric Range', 'Base MSRP', 'Model Year']], hue='Model Year', palette='bright')
plt.show()
```

Comprehensive Analysis with num_viz_analysis()
You can create a custom function num_viz_analysis() to automate visualizations for numerical features

Explanation: This function combines two visualizationshistogram and boxplotfor any given numerical column, making it easier to analyze multiple features efficiently.

```python
def num_viz_analysis(df, column):
    plt.figure(figsize=(14, 6))
    
    # Subplot 1: Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Histogram of {column}')
    
    # Subplot 2: Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
    
    plt.show()

# Call the function for 'Electric Range'
num_viz_analysis(df, 'Electric Range')
```

Boxplots are useful for quickly identifying outliers and understanding the distribution of a dataset. In this example:

Plot 1 (Electric Range): Reveals the electric range distribution and potential outliers in electric miles.
Plot 2 (DOL Vehicle ID): Visualizes the distribution of Vehicle IDs, though IDs might not have a clear statistical significance.
Plot 3 (Base MSRP): Analyzes the price range of vehicles, highlighting potential pricing outliers or anomalies.
